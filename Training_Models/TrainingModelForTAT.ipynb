# Data Slicing

```python
import matplotlib.pyplot as plt

def plot_graphs_separate(graph_points, model_name):
    sizes = list(graph_points['train_MSE'].keys())

    # Plot MSE
    plt.figure(figsize=(10, 5))
    plt.plot(sizes, list(graph_points['train_MSE'].values()), label='Train MSE', color='tab:red', linestyle='dashed')
    plt.plot(sizes, list(graph_points['val_MSE'].values()), label='Val MSE', color='tab:red')
    plt.xlabel('Test Set Size Ratio')
    plt.ylabel('MSE')
    plt.title(f'Mean Squared Error by Split for {model_name}')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Plot R²
    plt.figure(figsize=(10, 5))
    plt.plot(sizes, list(graph_points['train_R2'].values()), label='Train R²', color='tab:blue', linestyle='dashed')
    plt.plot(sizes, list(graph_points['val_R2'].values()), label='Val R²', color='tab:blue')
    plt.xlabel('Test Set Size Ratio')
    plt.ylabel('R² Score')
    plt.title(f'R² Score by Split for {model_name}')
    plt.legend()
    plt.grid(True)
    plt.show()

```

```python
def plot_score_and_loss_by_split_XGB(x, t):
    graph_points = {'train_MSE': {}, 'val_MSE': {}, 'train_R2': {}, 'val_R2': {}}
    
    scaler = preprocessing.MinMaxScaler()
    x = scaler.fit_transform(x)
    
    for size in range(10, 100, 10):
        x_train, x_val, t_train, t_val = model_selection.train_test_split(x, t, test_size=size/100, random_state=42)
      
        
        xgb_reg = XGBRegressor(objective='reg:squarederror', learning_rate=0.1, n_estimators=100, random_state=42)
        xgb_reg.fit(x_train, t_train)
        
        y_train_pred = xgb_reg.predict(x_train)
        y_val_pred = xgb_reg.predict(x_val)
        
        graph_points['train_MSE'][size/100] = metrics.mean_squared_error(t_train, y_train_pred)
        graph_points['val_MSE'][size/100] = metrics.mean_squared_error(t_val, y_val_pred)
        graph_points['train_R2'][size/100] = metrics.r2_score(t_train, y_train_pred)
        graph_points['val_R2'][size/100] = metrics.r2_score(t_val, y_val_pred)
    
    plot_graphs_separate(graph_points,model_name='XGBoost')

```

# XGBoosting

## Finding The Optimal Hyper-Parameters and Fine-Tuning

```python
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV

xgb = XGBRegressor(objective='reg:squarederror',random_state=42)

param_grid = {
    'learning_rate': [ 0.1],      # קצב למידה קטן עד בינוני
    'n_estimators': [100],         # מס
    'max_depth': [ 5],                  # עומק העץ
    'gamma': [0.1,0],                   # עונש על פיצול
    'reg_alpha': [ 0.496],                      # רגולריזציה L1
    'reg_lambda': [ 0.9],                   # רגולריזציה L2

}



grid_search = GridSearchCV(xgb, param_grid, scoring='neg_mean_squared_error', cv=5, verbose=2,
    n_jobs=-1)
grid_search.fit(X_train, t_train)

print("Best Parameters:", grid_search.best_params_)
```

```python
from sklearn.metrics import mean_squared_error

# Get the best model from the grid search
best_xgb = grid_search.best_estimator_

# Make predictions on the test set
y_pred = best_xgb.predict(X_train)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(t_train, y_pred)

print("Mean Squared Error:", mse)
```

```python
y_test_xg = best_xgb.predict(X_test)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(t_test, y_test_xg)
print("Mean Squared Error:", mse)
```

## Model Accuracy Evaluation

```python
# Define the ±16% accuracy criterion
tolerance = 0.15 * t_test  # 15% of actual values
correct_predictions = np.abs(y_test_xg - t_test) <= tolerance

# Calculate accuracy
accuracy = (np.sum(correct_predictions) / len(t_test)) * 100

# Print results
print(f"Model Accuracy (±15% threshold): {accuracy:.2f}%")
```

Although we set a target accuracy of 85%, the current model did not meet this threshold. Consequently, we will examine an alternative model , the Random Forest , to assess whether it can achieve better performance.



# RandomForest

## Data Slicing

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn import preprocessing, model_selection, metrics

def plot_score_and_loss_by_split_RF(x, t):
    graph_points = {'train_MSE': {}, 'val_MSE': {}, 'train_R2': {}, 'val_R2': {}}
    
    scaler = preprocessing.MinMaxScaler()
    x = scaler.fit_transform(x)
    
    for size in range(10, 100, 10):
        x_train, x_val, t_train, t_val = model_selection.train_test_split(
            x, t, test_size=size/100, random_state=42
        )
        
        rf_reg = RandomForestRegressor(n_estimators=100, random_state=42)
        rf_reg.fit(x_train, t_train)
        
        y_train_pred = rf_reg.predict(x_train)
        y_val_pred = rf_reg.predict(x_val)
        
        graph_points['train_MSE'][size/100] = metrics.mean_squared_error(t_train, y_train_pred)
        graph_points['val_MSE'][size/100] = metrics.mean_squared_error(t_val, y_val_pred)
        graph_points['train_R2'][size/100] = metrics.r2_score(t_train, y_train_pred)
        graph_points['val_R2'][size/100] = metrics.r2_score(t_val, y_val_pred)
    
    plot_graphs_separate(graph_points,model_name='Random Forest')
```

## Finding The Optimal Hyper-Parameters and Fine-Tuning

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error

# הגדרת רשת היפר-פרמטרים לחיפוש
param_grid = { 
    'n_estimators': [100, 200, 300,500],
    'max_depth': [5],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': [0.5,'sqrt', 'log2'],
    'bootstrap': [True, False]
}

# יצירת המודל
rf = RandomForestRegressor(random_state=42)

# חיפוש רשת עם 5-Fold Cross Validation
grid_search = GridSearchCV(rf, param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1, verbose=2)
grid_search.fit(X_train, t_train)

# התוצאות הטובות ביותר
best_rf = grid_search.best_estimator_
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)

# חיזוי ובדיקת הביצועים
y_pred_train = best_rf.predict(X_train)
y_pred_test = best_rf.predict(X_test)

mse_train = mean_squared_error(t_train, y_pred_train)
mse_test = mean_squared_error(t_test, y_pred_test)

print(f"MSE Train: {mse_train:.2f}, MSE Test: {mse_test:.2f}")
```

```python
# Make predictions on the test set
y_test_rf = best_rf.predict(X_test)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(t_test, y_test)

print("Mean Squared Error:", mse)
```

## Comparison between the models

```python
import numpy as np
import matplotlib.pyplot as plt

def compare_model_accuracies(model_predictions):
    accuracies = {}

    for model_name, (y_pred, y_true) in model_predictions.items():
        y_pred = np.array(y_pred)
        y_true = np.array(y_true)
        tolerance = 0.15 * y_true
        correct_predictions = np.abs(y_pred - y_true) <= tolerance
        accuracy = (np.sum(correct_predictions) / len(y_true)) * 100
        accuracies[model_name] = accuracy

    models = list(accuracies.keys())
    values = list(accuracies.values())

    plt.figure(figsize=(10, 6))
    bars = plt.bar(models, values, color='lightgreen', edgecolor='black')

    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, yval + 1, f'{yval:.2f}%', ha='center', fontsize=10)

    plt.ylabel('Accuracy (±15%)')
    plt.title('Model Accuracy Comparison (±15% Threshold)')
    plt.ylim(0, 110)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
